```python
import os
```

```python
# Load openai api key from .env file

from dotenv import load_dotenv
load_dotenv()
openapi_key = os.getenv('OPENAI_API_KEY')

# Check
print('openapi key:', bool(openapi_key))
```

```python
# Set up openai language model (LLM) 

from langchain_openai import OpenAI
openai_llm = OpenAI(temperature = 0.3, openai_api_key = openapi_key)
```

### Testing the LLM with a simple prompt

```python
# Example
output = openai_llm.invoke("What is Prompt Engineering?")
print(output)
```

```python
openai_llm = OpenAI(temperature = 0.7, openai_api_key = openapi_key)

output = openai_llm.invoke('What’s the difference between generative AI and traditional AI?')
print(output)
```

### Zero-shot prompting

```python
# Example

#  "Classify the text into neutral, negative, or positive."
prompt = 'Classify the sentiment: I love this product!'

model_output = openai_llm.invoke(prompt)

print('Output at Zero-shot prompting:\n', model_output)
```

### Few-shot prompting

```python
# Sentiment analysis using few-shot prompting

# Example

prompt = '''

Predict sentiment label:

Input: This is great news for hard working family!
Classification: Positive

Input: I don't like this environment. 
Classification: Negative

Input: It is fine for me.
Classification: Neutral

Input: The movie was awesome.
Classification:

'''
```

```python
result = openai_llm.invoke(prompt)
Classification
print("LLM output:\n", result)
```

```python
# Sentiment analysis with few-shot prompting focused on negative sentiment

# Example

prompt = '''
Predict sentiment label:

Input: I am not happy with this product.
Classification: Negative

Input: I had a bad experience.
Classification: Negative

Input: Very disappointing with this score.
Classification: Negative

Input: The movie was slow.
Classification:

'''
```

```python
llm_result = openai_llm.invoke(prompt)

print('LLM output:\n', llm_result)
```

### Chain-of-Thought (CoT) prompting used to guide the LLM step by step thinking to get top results

```python
# CoT Prompting

# Example

prompt = '''
Explain solution step by step:

A graduate student has 7 Physics textbooks. He buys 4 reference books. He gives 3 textbooks to his friend.  
How many books does he have now? You need to show all steps with reasoning.
'''

llm_result = openai_llm.invoke(prompt)

print('Chain of thought output:\n', llm_result)
```

### Python Code Generation

```python

prompt = '''
Write a Python function which sorts a list of numbers in descending order.
'''
llm_result = openai_llm.invoke(prompt)

print('Code Prompt output:\n', llm_result)
```
